# Matplotlib Example Workflow

```python
#O. import matplotlob and load it in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

#2. Setup Plot
fig, ax = plt.subplots(figsize=(10, 10)) #x and y dimensions of plot, width and height

# 3. Plot data
ax.plot(x, y)

# 4. Customize Plot
ax.set(title="Simple Plot", xlabel="x-axis", ylabel="y-axis")

#5. save and show (save the whole fig)
fig.savefig("images/sample-plot.png")
```

# Subplot option 1

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

#Plot each data
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000))
```

# Option 2 Subplot

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

#plot different ax
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000))
```

# Lets try on another data set

```python
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
#histogram for distribution on age col
heart_disease["age"].plot.hist(bins=50)
```

```python
heart_disease.head()
```

```python
heart_disease.plot.hist(figsize=(10, 30),subplots=True)
```

```python
# Which one should u use, (pyplot or matplot lib oo method)
```

```python
#quick pyplot is okay
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
over_50.plot(kind="scatter",
             x="age",
            y="chol",
            c="target") #Pyplot
```

```python
#oo method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax)
ax.set_xlim([45, 100])
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
#plot the data
scatter = ax.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])
ax.set(title="Heart Disease and Cholestrol Levels", xlabel="Age", ylabel ="Cholestrol")
ax.legend(*scatter.legend_elements(), title="Target") #legend
ax.axhline(y=over_50["chol"].mean(), linestyle="--") #horizontal line
```

```python
over_50
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharex=True)
scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])
ax0.set(title="Heart Diseases and Cholestrol evels", ylabel="cholestrol")
ax0.legend(*scatter.legend_elements(), title="Target")
ax0.axhline(y=over_50["chol"].mean(), linestyle="--")

catter = ax1.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"])
ax1.set(title="Heart Disease over Max heart rate", xlabel="age", ylabel="Thalach")
ax1.legend(*scatter.legend_elements(), title="Target")
ax1.axhline(y=over_50["thalach"].mean(), linestyle="--")
fig.suptitle("Heart Disease")
```

```python
plt.style.use("seaborn-whitegrid")
fig, ax = plt.subplots(figsize=(10, 6))
#plot the data
scatter = ax.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"], cmap ="winter") #changes color scheme
ax.set(title="Heart Disease and Cholestrol Levels", xlabel="Age", ylabel ="Cholestrol")
ax.legend(*scatter.legend_elements(), title="Target") #legend
ax.axhline(y=over_50["chol"].mean(), linestyle="--") #horizontal line
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharex=True)
scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"], cmap="winter")
ax0.set(title="Heart Diseases and Cholestrol evels", ylabel="cholestrol")
ax0.set_xlim(50, 80) #set lim
ax0.legend(*scatter.legend_elements(), title="Target")
ax0.axhline(y=over_50["chol"].mean(), linestyle="--")

catter = ax1.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"], cmap="winter")
ax1.set(title="Heart Disease over Max heart rate", xlabel="age", ylabel="Thalach")
ax1.set_xlim(50, 80)
ax1.set_ylim(60, 200)
ax1.legend(*scatter.legend_elements(), title="Target")
ax1.axhline(y=over_50["thalach"].mean(), linestyle="--")
fig.suptitle("Heart Disease")
```
